```python
def calc_degree(merged_table, id):
    degree_out = len(merged_table[merged_table['Id_Start'] == id])
    degree_in = len(merged_table[merged_table['Id_Target'] == id])
    return degree_in + degree_out
```

```python
sample_hspecs = list(hspec_stat[hspec_stat >= 5].index)
sample_hspecs
```

```python
stat_result = []
same_hspec_result = []
for i in range(100):
    sample_stat = dict([('sample_time', i)] + [(hspec,0) for hspec in sample_hspecs])
    input_sequence_sampled = pd.concat([data[~data['Host'].isin(sample_hspecs)] ,pd.concat([ hspecdf.sample(n=7) for hspec, hspecdf in data[data['Host'].isin(sample_hspecs)].groupby('Host')])])
    
    edge_table_sampled = edge_table[(edge_table['Source'].isin(input_sequence_sampled['Accession'])) & (edge_table['Target'].isin(input_sequence_sampled['Accession']))]
    merged_df_sampled = pd.merge(edge_table_sampled, node_table, left_on='Source', right_on='Id', how='left')
    merged_df_sampled = pd.merge(merged_df_sampled, node_table, left_on='Target', right_on='Id', how='left', suffixes=('_Start', '_Target'))
    #merged_df_sampled = merged_df_sampled[merged_df_sampled['Relation_Type']!='Recombination']
    merged_df_sampled_samehost = merged_df_sampled[merged_df_sampled['Host_Start'] == merged_df_sampled['Host_Target']]
    
    nodes = input_sequence_sampled[['Accession', 'Host']].set_index("Accession").to_dict()['Host']
    edges = [ (record['Source'], record['Target']) for index, record in merged_df_sampled.iterrows() ]
    group_sizes = {}
    for node, group in nodes.items():
        if group not in group_sizes:
            group_sizes[group] = 0
        group_sizes[group] += 1

    # 计算总节点数  
    total_nodes = len(nodes)

    # 计算观测的组内边数量
    observed_intra_edges = 0
    for edge in edges:
        if nodes[edge[0]] == nodes[edge[1]]:
            observed_intra_edges += 1

    # 计算期望的组内边数量
    expected_intra_edges = 0
    for group_size in group_sizes.values():
        if group_size > 1:
            # 计算组内任意两节点连接的概率
            prob = (group_size - 1) / (total_nodes - 1)
            # 所有可能的边中组内边的期望数量
            expected_intra_edges += prob * len(edges)

    expected_inter_edges = 0
    for edge in edges:
        node1, node2 = edge
        group1 = nodes[node1]
        group2 = nodes[node2]
        # 确保仅计算组间边
        if group1 != group2:
            # 该边是组间边
            expected_inter_edges += 1


    # 进行统计检验
    p_value = binom_test(observed_intra_edges, len(edges), expected_intra_edges / len(edges))

    
    same_hspec_result.append([i, total_nodes, len(merged_df_sampled),  observed_intra_edges, expected_intra_edges, expected_inter_edges, prob, p_value])
    
    
    recombination_accessions = set(list(merged_df_sampled['Id_Start']) + list(merged_df_sampled['Id_Target']))


    for hspec, hspecdf in input_sequence_sampled[input_sequence_sampled['Host'].isin(sample_hspecs)].groupby('Host'):
        sum_degree = 0
        for seqid in hspecdf['Accession']:
            sum_degree += calc_degree(merged_df_sampled, seqid)
        sample_stat[hspec] = sum_degree
    stat_result.append(sample_stat)
        
pd.DataFrame(stat_result).to_csv('degree_sampled_to_5seqs_v2.csv', index=False)
```
