```python
import pandas as pd
```

```python
distresult = pd.read_csv('bat_sarbecovirus_anicalc.csv').drop('vsname_abbr', axis=1)
distresult['mismatch_rate'] = distresult['mismatch'] / distresult['length']
distresult
```

```python
metadata = pd.read_csv('/jdfssz1/ST_HEALTH/P20Z10200N0206/wangdaxi/Projects/SouthChinaBat/Sarbecovirus/01.Sarbecovirus_genome_set/Rhinolophus_final_metatable.csv', header=None, names=['seqid', 'seqheader', 'province', 'species'])
metadata
```

```python
from Bio import AlignIO
import numpy as np
```

```python
alignment_file = "/jdfssz1/ST_HEALTH/P20Z10200N0206/wangdaxi/Projects/SouthChinaBat/Sarbecovirus/01.Sarbecovirus_genome_set/bat_sarbecovirus_cdhit_mafft_output_selected_Rhinolophus.fa"
alignment = AlignIO.read(alignment_file, "fasta")

num_populations = alignment.get_alignment_length()
num_sites = len(alignment)

# Convert the alignment into a numpy array for easier manipulation
alignment_array = np.array([list(rec.seq) for rec in alignment])

# Calculate allele frequencies for each population
allele_frequencies = np.sum(alignment_array, axis=0) / (2 * num_sites)

# Calculate Ht (total genetic variation)
ht = np.var(allele_frequencies)

hs = np.mean(np.var(alignment_array, axis=1))

# Calculate Fst
fst = (ht - hs) / ht
```

```python
lineage_info = pd.read_table('/jdfssz1/ST_HEALTH/P20Z10200N0206/renzirui/Projects/GIZ_SouthernChinaBat_clean/BatSarbecovirus/order_in_plot_given_ordering.txt', sep='\s+')
lineage_info['Lineage'] = lineage_info['Lineage'].astype(str)
metadata_addlineage = metadata.merge(lineage_info[['Name', 'Lineage']], right_on='Name', left_on='seqid', how='right').drop('Name', axis=1)
metadata_addlineage
```

```python
merged_info = distresult.merge(metadata_addlineage, left_on='sn1', right_on='seqid', how='left').merge(metadata_addlineage, left_on='sn2', right_on='seqid', how='left', suffixes=['_1', '_2'])
merged_info
```

```python
merged_info['same_province'] = merged_info['province_1'] == merged_info['province_2']
merged_info['same_species'] = merged_info['species_1'] == merged_info['species_2']
merged_info['same_lineage'] = merged_info['Lineage_1'] == merged_info['Lineage_2']
```

```python
merged_info
```

```python
province_fst = []
for province, _province_seqinfo in metadata.groupby('province'):
    province_distinfo = merged_info[(merged_info['province_1']==province) | (merged_info['province_2']==province)].copy()
    same_province = province_distinfo.loc[province_distinfo['same_province'], 'mismatch_rate'].mean()
    diff_province = province_distinfo.loc[~province_distinfo['same_province'], 'mismatch_rate'].mean()
    try:
        fst = (diff_province - same_province)/diff_province
    except:
        fst = 'Inf'
    province_fst.append([province, same_province, diff_province, fst])
```

```python
province_fstdf = pd.DataFrame(province_fst, columns=['province', 'mean_within_diversity', 'mean_between_diversity', 'estimated_Fst'])
```

```python
lineage_fst = []
for lineage, _lineage_seqinfo in metadata_addlineage.groupby('Lineage'):
    lineage_distinfo = merged_info[(merged_info['Lineage_1']==lineage) | (merged_info['Lineage_2']==lineage)].copy()
    same_lineage = lineage_distinfo.loc[lineage_distinfo['same_lineage'], 'mismatch_rate'].mean()
    diff_lineage = lineage_distinfo.loc[~lineage_distinfo['same_lineage'], 'mismatch_rate'].mean()
    try:
        fst = (diff_lineage - same_lineage) / diff_lineage
    except:
        fst = 'Inf'
    lineage_fst.append([lineage, same_lineage, diff_lineage, fst])
```

```python
lineage_fstdf = pd.DataFrame(lineage_fst, columns=['lineage', 'mean_within_diversity', 'mean_between_diversity', 'estimated_Fst'])
```

```python
with pd.ExcelWriter('Sarbecovirus_estimatedFst.xlsx') as excelwriter:
    province_fstdf.to_excel(excelwriter, index=False, sheet_name='Province')
    lineage_fstdf.to_excel(excelwriter, index=False, sheet_name='Lineage')
```
