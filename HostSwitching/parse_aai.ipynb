```python
import pandas as pd
import re
```

```python
kept_chunks = []
for chunk in pd.read_csv('5.aai_result/CoV_aai.csv', chunksize=1e6):
    _filtered_chunk = chunk[(chunk['pident'] >= 75) & (chunk['gap']< 50) & (chunk['alnlen'] > 180)]
    _suffix_lines_seq1 = _filtered_chunk['seq1'].apply(lambda x: bool(re.search(r'.*_\d$', x)))
    _filtered_chunk.loc[_suffix_lines_seq1, 'seq1'] = _filtered_chunk.loc[_suffix_lines_seq1, 'seq1'].apply(lambda x: '_'.join(x.split('_')[:-1]))
    _suffix_lines_seq2 = _filtered_chunk['seq2'].apply(lambda x: bool(re.search(r'.*_\d$', x)))
    _filtered_chunk.loc[_suffix_lines_seq2, 'seq2'] = _filtered_chunk.loc[_suffix_lines_seq2, 'seq2'].apply(lambda x: '_'.join(x.split('_')[:-1]))
    kept_chunks.append(_filtered_chunk)
```

```python
filtered_aaidf = pd.concat(kept_chunks)
filtered_aaidf
```

```python
virioninfo_human = pd.read_table('/jdfssz1/ST_HEALTH/P20Z10200N0206/wangdaxi/Projects/SouthChinaBat/host_switching/RepViralSequences_133seqs_human.blast.add', skiprows=1, header=None, 
                           names='qseqid sseqid pident length mismatch gapopen qstart qend sstart send evalue bitscore host_species virusname host_species2'.split())
virioninfo_nonhuman = pd.read_table('/jdfssz1/ST_HEALTH/P20Z10200N0206/wangdaxi/Projects/SouthChinaBat/host_switching/RepViralSequences_133seqs_nonhuman.blast.add', skiprows=1, header=None, 
                           names='qseqid sseqid pident length mismatch gapopen qstart qend sstart send evalue bitscore host_species virusname host_species2'.split())
virioninfo_full = pd.concat([virioninfo_human, virioninfo_nonhuman])
```

```python
virioninfo_full
```

```python
virioninfo_full_to_merge = virioninfo_full[['sseqid', 'host_species', 'virusname']].drop_duplicates()
virioninfo_full_to_merge
```

```python
filtered_aaidf_mergeinfo = filtered_aaidf.merge(virioninfo_full_to_merge.rename(columns={"sseqid": "seq1", "host_species":"hspec1", "virusname":"vname1"}), how='left', on='seq1') \
    .merge(virioninfo_full_to_merge.rename(columns={"sseqid": "seq2", "host_species":"hspec2", "virusname":"vname2"}), how='left', on='seq2')
```

```python
filtered_aaidf_mergeinfo.dropna().sort_values('pident', ascending=False).drop_duplicates(['hspec1', 'vname1', 'hspec2', 'vname2'])
```
