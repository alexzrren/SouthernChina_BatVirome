```python
import pandas as pd
```

```python
blastresult = pd.read_table('sco_blastn.m8', header=None, names='qseqid qlen saccver slen qstart qend sstart send evalue bitscore length pident mismatch gaps'.split())
blastresult['saccver'] = 'lcl|' + blastresult['saccver']
blastresult = blastresult[blastresult['length']>=100]
```

```python
filtered_pairwise = blastresult.sort_values('bitscore', ascending=False).drop_duplicates('qseqid').drop_duplicates('saccver')
filtered_pairwise
```

```python
Ra_validcov_cdslist = open('../../coverage/Ra.sufficient_coverage.CDSlist.tsv').read().splitlines()
Rs_validcov_cdslist = open('../../coverage/Rs.sufficient_coverage.CDSlist.tsv').read().splitlines()
```

```python
Rs_validcov_cdslist
```

```python
final_kept_pairwise = filtered_pairwise[(filtered_pairwise['qseqid'].isin(Ra_validcov_cdslist)) & (filtered_pairwise['saccver'].isin(Rs_validcov_cdslist))]
final_kept_pairwise['startpos'] = 0
final_kept_pairwise
```

```python
final_kept_pairwise[['qseqid', 'startpos', 'qlen']].to_csv('Ra_keptSCO_CDS.bed', header=None, index=None, sep='\t')
final_kept_pairwise[['saccver', 'startpos', 'slen']].to_csv('Rs_keptSCO_CDS.bed', header=None, index=None, sep='\t')
```
